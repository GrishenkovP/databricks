```
%sql
select o.order_id, 
       o.order_status, 
       c.customer_unique_id, 
       c.customer_city, 
       c.customer_state,
       op.payment_type,
       round(op.payment_value,0) as payment_value
from orders as o left join customers as c on o.customer_id = c.customer_id
                 left join order_payments as op on o.order_id = op.order_id
```

```
%sql
select o.order_id, 
       o.order_status, 
       c.customer_unique_id, 
       c.customer_city, 
       c.customer_state,
       op.payment_type,
       round(op.payment_value,0) as payment_value
from orders as o left join customers as c on o.customer_id = c.customer_id
                 left join order_payments as op on o.order_id = op.order_id
where (op.payment_type is not null) and (op.payment_type != "not_defined")
```

```
%sql
select o.order_id, 
      o.order_status, 
      c.customer_unique_id, 
      c.customer_city, 
      c.customer_state,
      op.payment_type,
      round(op.payment_value,0) as payment_value
from orders as o left join customers as c on o.customer_id = c.customer_id
                 left join order_payments as op on o.order_id = op.order_id
where (op.payment_type is not null) and (op.payment_type != "not_defined")
```

```
%sql
select oi.order_id,
       oi.price,
       oi.freight_value,
       round((oi.price + oi.freight_value),0) as total_value,
       pc.product_category_name_english as product_category,
       s.seller_city,
       s.seller_state
from order_items as oi left join products as p on oi.product_id = p.product_id
                       left join product_category_name_translation as pc on p.product_category_name = pc.product_category_name
                       left join sellers as s on oi.seller_id = s.seller_id
```

```
%sql
with tbl as (select oi.order_id,
       oi.price,
       oi.freight_value,
       round((oi.price + oi.freight_value),0) as total_value,
       pc.product_category_name_english as product_category,
       s.seller_city,
       s.seller_state
from order_items as oi left join products as p on oi.product_id = p.product_id
                       left join product_category_name_translation as pc on p.product_category_name = pc.product_category_name
                       left join sellers as s on oi.seller_id = s.seller_id)
                       
select t.product_category,
       sum(t.total_value) as total_value
from tbl as t
group by t.product_category
order by sum(t.total_value) desc
limit 10
```
