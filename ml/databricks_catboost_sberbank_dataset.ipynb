```
# Импорт библиотек
import pandas as pd
import numpy as np
```

### Загрузка датасета

```
df = spark.read.format('parquet').load('/tmp/dataset/data.parquet').toPandas()
```

```
df.head()
```

### Обработка столбцов с датами

```
df['BIRTH_DATE'] = (np.datetime64('2018-01-01') + df['DAYS_BIRTH'].astype(np.dtype('timedelta64[D]'))).astype(str)
df['EMP_DATE'] = (np.datetime64('2018-01-01') + np.clip(df['DAYS_EMPLOYED'], None, 0).astype(np.dtype('timedelta64[D]'))
                    ).astype(str)
df.drop(['DAYS_BIRTH', 'DAYS_EMPLOYED'], axis=1, inplace=True)
```

### Приведение наименований столбцов в датафрейме к нижнему регистру

```
df.columns = [_.lower() for _ in df.columns]
```

### Замена пропущенных значений в датафрейме

```
for col_name in df.columns:
  if (df[col_name].dtype == np.dtype('int32')) or (df[col_name].dtype == np.dtype('float64')):
    df[col_name].fillna(0, inplace=True)
  else:
    df[col_name].fillna('-', inplace=True)
```

```
df.isnull().sum()
```

```
df.head()
```

### Проверка сбалансированности классов (очень сильный перекос в пользу 0!)

```
df['target'].value_counts(normalize=True)
```

### Определение названий столбцов с числовами и текстовыми значениями

```
X = df.drop('target', axis=1)
y = df.target
```

```
cat_features = []

for col_name in df.columns:
  if (df[col_name].dtype not in [np.dtype('int32'), np.dtype('float64')]):
    cat_features.append(col_name)
```

```
print(cat_features)
```

### Разбиение на тренировочный и тестовый датасет

```
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
TEST_SIZE = 0.4
TARGET_NAME = 'target' # Target column name

X_train, X_validation, y_train, y_validation = train_test_split(X, y, 
                                                                    test_size=TEST_SIZE, 
                                                                    stratify=df[TARGET_NAME], 
                                                                    random_state=RANDOM_STATE)
```

### Обучение модели

```
from catboost import CatBoostClassifier
from sklearn.utils.class_weight import compute_class_weight

classes = np.unique(y_validation)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = dict(zip(classes, weights))


model = CatBoostClassifier(verbose=100,
                           random_state=RANDOM_STATE,
                           learning_rate=0.05,
                           early_stopping_rounds=200,
                           eval_metric='AUC',
                          cat_features=cat_features,
                           class_weights=class_weights
                           )
```

```
model.fit(X_train,y_train, eval_set=(X_validation,y_validation))
```

### Оценка точности модели

```
from sklearn import metrics

predictions = model.predict(X_validation)
print(metrics.classification_report(y_validation, predictions))
```
