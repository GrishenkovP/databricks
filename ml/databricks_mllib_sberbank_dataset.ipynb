### Подготовка данных

```
df = spark.read.format('parquet').load('/tmp/dataset/data.parquet')
```

```
df.limit(10).toPandas().head()
```

```
import pyspark.sql.functions as F
from pyspark.sql.functions import date_add

# Переводим даты в формат понятный человеку
df = df.withColumn('DATE_START', F.lit('2018-01-01'))
df = df.withColumn('BIRTH_DATE', F.expr('date_add(DATE_START, DAYS_BIRTH)'))
df = df.withColumn('EMP_DATE', F.expr('date_add(DATE_START, DAYS_EMPLOYED)'))
columns_to_drop = ['DAYS_BIRTH', 'DAYS_EMPLOYED','DATE_START']
df = df.drop(*columns_to_drop)
```

```
# Переводим даты в unix формат для машинного обучения
df = df.withColumn('BIRTH_DATE_UNIX', F.unix_timestamp(df.BIRTH_DATE, 'yyyy-MM-dd'))
df = df.withColumn('EMP_DATE_UNIX', F.unix_timestamp(df.EMP_DATE, 'yyyy-MM-dd'))
df = df.drop('BIRTH_DATE').drop('EMP_DATE')
```

```
# Заполняем пусты в датасете
df = df.na.fill(0)
df = df.na.fill('-')
```

```
df.limit(10).toPandas().head()
```

```
df.printSchema()
```

### Разбивка на тренировочный и тестовый датасет

```
RANDOM_STATE = 42
TEST_SIZE = 0.4
TRAIN_SIZE = 0.6

trainDF, testDF = df.randomSplit([TRAIN_SIZE, TEST_SIZE], seed=RANDOM_STATE)
print(f'''Размер первоначального датасета - {df.count()}. Тренирочный массив данных -  {trainDF.count()}, проверочный -  {testDF.count()}''')
```

### Кодировка

```
from pyspark.ml.feature import OneHotEncoder, StringIndexer, VectorAssembler


categoricalCols = [field for (field, dataType) in trainDF.dtypes if dataType == 'string']
indexOutputCols = [x + 'Index' for x in categoricalCols]
oheOutputCols = [x + 'OHE' for x in categoricalCols]

stringIndexer = StringIndexer(inputCols=categoricalCols, outputCols=indexOutputCols, handleInvalid='skip')
oheEncoder = OneHotEncoder(inputCols=indexOutputCols, outputCols=oheOutputCols)

numericCols = [field for (field, dataType) in trainDF.dtypes if ((dataType in ['double', 'integer', 'long']) & (field != 'TARGET'))]
assemblerInputs = oheOutputCols + numericCols
vecAssembler = VectorAssembler(inputCols=assemblerInputs, outputCol='features')
```

### Построение пайплайна с обучением и прогнозом

```
from pyspark.ml import Pipeline
from pyspark.ml.classification import RandomForestClassifier


rf = RandomForestClassifier(labelCol='TARGET', numTrees=20, maxDepth=5, maxBins=32, impurity='gini')
pipeline = Pipeline(stages = [stringIndexer, oheEncoder, vecAssembler, rf])

pipelineModel = pipeline.fit (trainDF)
predDF = pipelineModel.transform(testDF)
predDF.select('TARGET','prediction').show()
```

### Оценка качества работы алгоритма

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator_accuracy = MulticlassClassificationEvaluator(labelCol='TARGET', predictionCol='prediction', metricName='accuracy')
accuracy = evaluator_accuracy.evaluate(predDF)
print('Accuracy = ' + str(accuracy))
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluator_areaunderroc = BinaryClassificationEvaluator(labelCol='TARGET', metricName='areaUnderROC') 
areaunderroc = evaluator_areaunderroc.evaluate(predDF)
print('AUC = ' + str(areaunderroc))
```

```
from sklearn import metrics

t = predDF.select('TARGET','prediction').toPandas()
print(metrics.classification_report(t.TARGET, t.prediction))
```
